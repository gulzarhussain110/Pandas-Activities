```python
# creat function to know k is batsman na kis team k against high run score kiya ha.
def top_runs_against(batsman):
    batsman=data['batsman']==batsman
    data[batsman]
    bowlingteam=data[batsman].groupby('bowling_team')
    return bowlingteam['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
top_runs_against('V Kohli')
```

```python
# sirf runs pta karnay hy tou values
def top_runs_against(batsman):
    batsman=data['batsman']==batsman
    data[batsman]
    bowlingteam=data[batsman].groupby('bowling_team')
    return bowlingteam['batsman_runs'].sum().sort_values(ascending=False).values[0]
```

```python
top_runs_against('V Kohli')
```

```python
# agar sirf team ka pat karna hy tou sirf index liknay
def top_runs_against(batsman):
    batsman=data['batsman']==batsman
    data[batsman]
    bowlingteam=data[batsman].groupby('bowling_team')
    return bowlingteam['batsman_runs'].sum().sort_values(ascending=False).index[0]
```
