```python
# creat function to know k is batsman na kis team k against high run score kiya ha.
def top_runs_against(batsman):
    batsman=data['batsman']==batsman
    data[batsman]
    bowlingteam=data[batsman].groupby('bowling_team')
    return bowlingteam['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
# agar sirf team ka pat karna hy tou sirf index liknay
def top_runs_against(batsman):
    batsman=data['batsman']==batsman
    data[batsman]
    bowlingteam=data[batsman].groupby('bowling_team')
    return bowlingteam['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
#Question sb sy khatarnak strike rate kis batsman ka hy 16 sy 20 overs k darmiyan
#srtike rate =(no of runs/no of balls)/100
```

```python
# sb sy pehlay over 16 sy 20 tk ka record nikalay gay
DeathOverData=data['over']>=16
```

```python
# wo saraybatsman aagay jinho na 16 sy 20 k overs may kitnay run kiya ha == count sy pta chalay ga k kis na kitnay balls paly ki hy
data[DeathOverData].groupby('batsman')['batsman_runs'].count()
```

```python
# agar mujhya ya condition lagani hay k kam sy kam 200 balls kahli ho
# upar wo saray batsman thay jinho na 1 khali ho ya 200 balls khali ho
all_batsman=data[DeathOverData].groupby('batsman')['batsman_runs'].count()
```

```python
# ab may na us data py condition laga di k kam sy kam 200
x=all_batsman>200
# ab all_batsman wala data py mask kr diya filter wali condition
# is in function k liya zarori hota hay k data list ki form may ho is liya just players k name ko index sy fetch kar k list may convert kar diya ha
all_batsmanlist=all_batsman[x].index.tolist()
```

```python
data['batsman'].isin(all_batsmanlist)
```

```python
# ab masking karay gay ta k real data fetch ho sakay un batsman k hisab sy
masking=data['batsman'].isin(all_batsmanlist)
final=data[masking]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['ball'].count()
```

```python
strikeRate=(runs/balls)*100
```
